```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
fpath='/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week04/Data/weight-height male-female - sample.csv'
df = pd.read_csv(fpath)
```

```python
df.head()
```

- Creating a male and Female Filters

```python
female_filter = df['Gender']=='Female'
df_female = df.loc[female_filter]
df_female.head()
```

```python
df_male = df.loc[~female_filter]
df_male.head()
```

- Making 2 side-by-side subplots

```python
fig, axes = plt.subplots(ncols=2)
```

```python
fig, axes = plt.subplots(ncols=2,figsize=(10,5))
```

```python
fig, axes = plt.subplots(nrows=2,figsize=(10,5))
```

```python
fig, axes = plt.subplots(ncols=2,figsize=(10,5))
axes[0].set_title('axes[0]')
axes[1].set_title('axes[1]')
```

- Create a 2 side by side plots: male scatter plots of hight vs weight on the left, females on the right

```python
plt.style.use('dark_background')


fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], color = 'blue')
axes[0].set_title('Males', fontsize=12)
axes[0].set_ylabel('Weight(KG)')
axes[0].set_xlabel('Hight(cm)')

axes[1].scatter(df_female['Height'], df_female['Weight'], color = 'orange')
axes[1].set_title('Females', fontsize=12)
axes[1].set_ylabel('Weight(KG)')
axes[1].set_xlabel('Hight(cm)')


```

```python
plt.style.use('dark_background')

fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], color = 'blue')
axes[0].set_title('Males', fontsize=12)
axes[0].set_ylabel('Weight(KG)')
axes[0].set_xlabel('Hight(cm)')

axes[1].scatter(df_female['Height'], df_female['Weight'], color = 'orange')
axes[1].set_title('Females', fontsize=12)
axes[1].set_ylabel('Weight(KG)')
axes[1].set_xlabel('Hight(cm)')


fig.suptitle("Comparing Height Vs Weight for both Genders",fontsize=16)
```

```python

plt.style.use('dark_background')

fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], color = 'blue')
axes[0].set_title('Males', fontsize=12)
axes[0].set_ylabel('Weight(KG)')
axes[0].set_xlabel('Hight(cm)')

axes[1].scatter(df_female['Height'], df_female['Weight'], color = 'orange')
axes[1].set_title('Females', fontsize=12)
axes[1].set_ylabel('Weight(KG)')
axes[1].set_xlabel('Hight(cm)')


fig.suptitle("Comparing Height Vs Weight for both Genders",fontsize=16)
fig.get_tight_layout()
```

```python

plt.style.use('dark_background')

fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(10,4), sharex=True, sharey=True)

# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], color = 'blue')
axes[0].set_title('Males', fontsize=12)
axes[0].set_ylabel('Weight(KG)')
axes[0].set_xlabel('Hight(cm)')

axes[1].scatter(df_female['Height'], df_female['Weight'], color = 'orange')
axes[1].set_title('Females', fontsize=12)
axes[1].set_ylabel('Weight(KG)')
axes[1].set_xlabel('Hight(cm)')


fig.suptitle("Comparing Height Vs Weight for both Genders",fontsize=16)
fig.get_tight_layout()
```

```python
plt.style.use('dark_background')

fig, axes=plt.subplots(nrows=2, ncols=1, figsize=(6,8), sharex=True)

sns.scatterplot(data=df_male, x='Height', y='Weight', ax=axes[0], color='blue')
axes[0].set_title('Males', fontsize=12)
axes[0].set_ylabel('Weight(KG)')
axes[0].set_xlabel('Hight(cm)')


df_male['Height'].hist(bins='auto', edgecolor='black', ax=axes[1], color='blue')
axes[1].set_title('Histogram for Males')
axes[1].set_xlabel('Hight(Cm)')
axes[1].set_ylabel('Counts per Bin')

fig.tight_layout()
```

```python

plt.style.use('dark_background')

fig, axes=plt.subplots(nrows=2, ncols=2, figsize=(10,8), sharex=True)
# Subplot 1
axes[0,0].scatter(df_male['Height'], df_male['Weight'], color = 'blue')
axes[0,0].set_title('Males', fontsize=12)
axes[0,0].set_ylabel('Weight(KG)')
axes[0,0].set_xlabel('Hight(cm)')

axes[0,1].scatter(df_female['Height'], df_female['Weight'], color = 'orange')
axes[0,1].set_title('Females', fontsize=12)
axes[0,1].set_ylabel('Weight(KG)')
axes[0,1].set_xlabel('Hight(cm)')


df_male['Height'].hist(bins='auto', edgecolor='black', ax=axes[1,0], color='blue')
axes[1,0].set_title('Histogram for Males')
axes[1,0].set_xlabel('Hight(Cm)')
axes[1,0].set_ylabel('Counts per Bin')

df_female['Height'].hist(bins='auto', edgecolor='black', ax=axes[1,1], color='orange')
axes[1,1].set_title('Histogram for Females')
axes[1,1].set_xlabel('Hight(Cm)')
axes[1,1].set_ylabel('Counts per Bin')


fig.suptitle("2D Subplots Comparing Males Vs. Females",fontsize=16)
fig.get_tight_layout()
```
